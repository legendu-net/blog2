````python
def fix_code_blocks(path: Path) -> None:
    if path.suffix != ".md":
        return
    text = path.read_text()
    patterns = {
        r"```\n\s*:::bash\n": r"```bash\n",
        r"```\n\s*:::fish\n": r"```fish\n",
        r"```\n\s*:::sql\n": r"```sql\n",
        r"```\n\s*:::json\n": r"```json\n",
        r"```\n\s*:::text\n": r"```text\n",
        r"```\n\s*:::python\n": r"```python\n",
        r"```\n\s*:::rust\n": r"```rust\n",
    }
    for pat, rep in patterns.items():
        text = re.sub(pat, rep, text)
    path.write_text(text)
````

````python
for p in paths:
    fix_code_blocks(p)
````
