````python
import mdformat
````

````python
text = """
# My Title
```{admonition} Note
This is a MyST directive.
```

:::{list-table}
- - name
  - age
- - Jean Doe
  - 40
:::
"""
print(text)
````

````python
print(
    mdformat.text(
        text,
        extensions={"myst"},
        #codeformatters={"python", "rust"},
    )
)
````
