## Using Boolean Indexing

```python
import numpy as np
import pandas as pd

s = pd.Series([True, False, True, True, False, False, False, True])
s[s].index
```

Use `.values` if you need a `np.array` object.

```python
s[s].index.values
```

## Using np.nonzero

```python
np.nonzero(s)
```

## Using np.flatnonzero

```python
np.flatnonzero(s)
```

## Using np.where

```python
np.where(s)[0]
```

## Using np.argwhere

```python
np.argwhere(s).ravel()
```

## Using pd.Series.index

```python
s.index[s]
```

## Using python's built-in filter

```python
[*filter(s.get, s.index)]
```

## Using list comprehension

```python
[i for i in s.index if s[i]]
```
