## Unpacking Variables (Multiple Assignments)

You can do multiple assignments simialarly as what you can do in Python.

```rust
let t = ("how", 1);
```

```rust
t.0
```

```rust
t.1
```

```rust
let(v1, v2) = ("how", 1)
```

```rust
v1
```

```rust
v2
```
