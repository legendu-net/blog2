```python
import getpass
import requests
```

```python
token = getpass.getpass()
```

```python
requests.delete(
    url="https://api.github.com/repos/legendu-net/docker-jupyterhub/actions/secrets/GITHUBACTIONS",
    headers={"Authorization": f"token {token}"},
)
```
