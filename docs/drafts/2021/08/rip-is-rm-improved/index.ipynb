[rip2](https://github.com/MilesCranmer/rip2)
is the successor of rip.

## Installation

### cargo
```
cargo install --locked rip2
```
### icon
Or you can use the following command 
if you use the tool icon.
Notice that icon automatically configures rip
to add a symbolic link to `/usr/local/bin`.
```
icon rip2 -ic
```
### Homebrew
```
brew install rip2
```
