## Tips and Traps

1. Forggeting `$` is a common mistake when using a shell variable.

```java
%%bash

for x in 1 2 3 4 5; do
	echo $x
done
```

```java
%%bash

for x in {1..5}; do
	echo $x
done
```

```java
%%bash

for x in {1..5..2}; do
	echo $x
done
```

```java
%%bash

for x in $(seq 1 2 5); do
    echo $x
done
```

```java
%%bash

for (( i=1; i<=5; i++ )); do  
   echo $i
done
```

```java
%%bash

for (( i=1; i<=5; ++i )); do  
   echo $i
done
```

Below is an infinite loop.

```java
%%bash

for (( ; ; )); do  
   echo "This is a infinite loop!"
done
```

```java
%%bash

for dir in $(ls -d */); do
    echo $dir
done
```
